```python
import pandas as pd

def load_data(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        return "File not found"
    except pd.errors.ParserError:
        return "Error parsing the file"
    
df = load_data('data/Dataset.csv')
```
